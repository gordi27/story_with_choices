```python
# Словарь с предметами, необходимыми по сюжету
used_stuff = {'Вещи из дома': {'кепка', 'аптечка', 'фонарик', 'дротик'},
              'Найденная еда': ['лесные орехи', 'дикие яблоки', 'неизвестные ягоды']
              }

# Словарь, хранящий предметы, полученные в течение истории
belongings = {'Рюкзак': set(),  # Множество вещей в рюкзаке
              'Еда': []  # Список еды в инвентаре
              }

# Счетчик неприятностей, влияет на концовку
troubles = 0

# Имя файла для записи истории
filename = 'File1.txt'

# Очистка файла перед началом записи
with open(filename, "w", encoding="utf-8") as outfile:
    outfile.write("")


# Функция для вывода строки в консоль и записи в файл
def print_and_save(line):
    print(line)
    with open(filename, "a", encoding="utf-8") as outfile:
        print(line, file=outfile)


# Функция для записи в файл и возвращения ввода пользователя
def input_and_save(line):
    user_input = input(line)
    with open(filename, "a", encoding="utf-8") as outfile:
        print(f'Пользователь ввел: {user_input}', file=outfile)

    return user_input


# Функция для первой неприятности - палящее солнце
def sun():
    global troubles
    print_and_save('Ты на какое-то время вышел из леса, чтобы прогуляться вдоль поляны с цветами.')

    if 'кепка' not in belongings['Рюкзак']:  # Если у студента не оказалось кепки в рюкзаке
        print_and_save('Но солнце сегодня особенно яркое. Из-за отсутствия кепки ты получил солнечный удар!\n'
                       'Вот же неприятность... Если так и дальше продолжится, то придется возвращаться домой!')

        troubles += 1  # Счетчик неприятностей увеличился
        print_and_save(f'Счетчик неприятностей: {troubles}')
    else:
        print_and_save('Сегодня солнечно, но кепка уберегла тебя от перегрева!')


# Функция для второй неприятности - крапива в темноте
def nettle():
    global troubles
    print_and_save('Для продолжения пути пришлось пройти через небольшую, но самую густую часть леса, '
                   'куда не пробиваются солнечные лучи.')

    if 'фонарик' not in belongings['Рюкзак']:  # Если у студента не оказалось фонарика в рюкзаке
        print_and_save('В темноте без фонарика ты не замечаешь, как наступаешь в заросли крапивы! Это больно!')

        troubles += 1  # Счетчик неприятностей увеличился
        print_and_save(f'Счетчик неприятностей: {troubles}')
    else:
        print_and_save('Хорошо, что у тебя был фонарик! Чуть было не наступил в заросли крапивы.')


# Начало истории
print_and_save('Уставший от городской суеты, ты - студент, решил выбраться на природу в поход, чтобы увидеть водопад, '
               'про который ты много слышал.\n'
               'Проснувшись рано утром, ты собрал рюкзак и заметил, что в нем еще осталось немного места.\n'
               'Какие две вещи ты возьмешь с собой в поход?')

print_and_save(', '.join(used_stuff['Вещи из дома']))  # Вывод вещей в рюкзак на выбор

# Выбор двух предметов в рюкзак
while len(belongings['Рюкзак']) < 2:  # Пока в рюкзаке имеется место
    backpack_choice = input_and_save('Выбери два разных предмета (напиши через пробел): ')
    backpack_choice = backpack_choice.lower().split(' ')

    if len(backpack_choice) == 2 and (backpack_choice[0] != backpack_choice[1]):  # Проверка корректности ввода
        if (backpack_choice[0] in used_stuff['Вещи из дома'] and
                backpack_choice[1] in used_stuff['Вещи из дома']):  # Проверка корректности ввода
            for i in backpack_choice:
                print_and_save(f'+ {i}')  # Вывод выбранных вещей
                belongings['Рюкзак'].add(i)
        else:
            print_and_save('Выбери предметы из списка!')
    else:
        print_and_save('Выбери два разных предмета!')

# Выбор маршрута
route_choice = ''
print_and_save('Впереди тебя ждет выбор между длинным изнуряющим маршрутом и коротким, но опасным.')
while route_choice not in ['к', 'д']:  # проверка корректности ввода
    route_choice = input_and_save('Какой маршрут выберешь? Введи ответ в формате (к/д): ').lower()

sun()  # Первая неприятность на пути
nettle()  # Вторая неприятность

flag = True  # Флаг для определения завершенности истории
if troubles < 2:  # Если неприятностей не слишком много, продолжаем путь
    if route_choice == 'д':  # Если выбрана длинная дорога
        print_and_save('Вокруг тебя высокие деревья, пахнет свежестью, слышен легкий шелест листьев.\n'
                       'Ты наслаждаешься своим маленьким путешествием.')
        print_and_save('Продолжая свой путь, ты решил исследовать окрестности.')
        for food in used_stuff['Найденная еда']:  # Студент подряд находит разную еду
            print_and_save(f'Ты нашел {food}.')

            while True:   # Проверка корректности ввода
                choice = input_and_save(f'Съесть {food} сейчас? (да/нет): ').lower()
                if choice in ('да', 'нет'):
                    break
                else:
                    print_and_save('Введи ответ в формате (да/нет)')

            if choice == 'нет':
                print_and_save(f'+ {food}')  # Если еда не была съедена сразу, добавляем ее в инвентарь студента
                belongings['Еда'].append(food)

            elif food == 'неизвестные ягоды' and choice == 'да':  # Если неизвестные ягоды были съедены
                print_and_save('Сначала вкус кажется приятным, но внезапно твое тело слабеет, '
                               'голова начинает кружиться.')
                if 'аптечка' in belongings['Рюкзак']:  # Если в рюкзаке была аптечка с противоядием
                    print_and_save('Хорошо, что у тебя была аптечка. Ты нашел в ней противоядие.')

                else:
                    # Концовка: Студент отравился ягодами.
                    # Аптечки не оказалось, пришлось возвращаться в город.
                    print_and_save('Перед глазами все плывет. Аптечки в рюкзаке не оказалось, так что '
                                   'нужно возвращаться в город и обратиться к врачу.')
                    print_and_save('Конец!')
                    flag = False  # Конец истории

        if flag is True:  # История еще не завершилась, продолжаем
            print_and_save('Тебя окружает красота дикой природы. На удивление, здесь очень тихо.\n'
                           'Так тихо, что слышно урчание твоего живота.')
            if len(belongings['Еда']) == 0:
                # Концовка: У студента не оказалось еды. Он ослаб и ему пришлось возвращаться домой.
                print_and_save('До водопада оставалось совсем немного, но у тебя не осталось еды, '
                               'чтобы восстановить силы и идти дальше.\n'
                               'Ты возвращаешься домой.')
                print_and_save('Конец!')

            elif len(belongings['Еда']) == 1:  # Если остался один вид еды
                chosen_food = belongings['Еда'][0]
                print_and_save(f'Ты съедаешь {chosen_food}.')
                if chosen_food == 'неизвестные ягоды':  # Если были выбраны ягода
                    print_and_save('Сначала вкус кажется приятным, но внезапно твое тело слабеет, '
                                   'голова начинает кружиться.')
                    if 'аптечка' in belongings['Рюкзак']:  # Если в рюкзаке была аптечка с противоядием
                        # Концовка: Студент восстановил силы и успешно завершил поход.
                        print_and_save('Хорошо, что у тебя была аптечка. Ты нашел в ней противоядие.\n'
                                       'Ты восстановил свои силы и успешно доходишь до водопада!')
                        print_and_save('Конец!')
                    else:
                        # Концовка: Студент отравился ягодами.
                        # Аптечки не оказалось, пришлось возвращаться в город.
                        print_and_save('Перед глазами все плывет. Аптечки в рюкзаке не оказалось, так что '
                                       'нужно возвращаться в город и обратиться к врачу.')
                        print_and_save('Конец!')
                        flag = False

                else:  # Если были выбраны орехи или яблоки
                    # Концовка: Студент восстановил силы и успешно завершил поход.
                    print_and_save('Ты восстановил силы и успешно доходишь до водопада!')
                    print_and_save('Конец!')

            else:  # Если осталось несколько видов еды
                for i in range(len(belongings['Еда'])):
                    food = belongings['Еда'][i]
                    print_and_save(f'{(i + 1)}. {food}')  # Пронумерованный вывод еды на выбор
                while True:  # Проверка корректности ввода
                    # Выбор еды для перекуса с помощью номера
                    choice = input_and_save('Что ты съешь? Введи номер: ')
                    if choice.isdigit():  # Проверка корректности ввода
                        choice = int(choice)
                        if 0 < choice <= len(belongings['Еда']):  # Проверка корректности ввода
                            chosen_food = belongings['Еда'][choice - 1]
                            print_and_save(f'Ты съедаешь {chosen_food}')
                            break
                        else:
                            print_and_save('Введи корректный номер!')
                    else:
                        print_and_save('Введи число!')

                if chosen_food == 'неизвестные ягоды':  # Если были выбраны ягода
                    print_and_save('Сначала вкус кажется приятным, но внезапно твое тело слабеет, '
                                   'голова начинает кружиться.')
                    if 'аптечка' in belongings['Рюкзак']:  # Если в рюкзаке была аптечка с противоядием
                        # Концовка: Студент восстановил силы и успешно завершил поход.
                        print_and_save('Хорошо, что у тебя была аптечка. Ты нашел в ней противоядие.\n'
                                       'Ты восстановил свои силы и успешно доходишь до водопада!')
                        print_and_save('Конец!')
                    else:
                        # Концовка: Студент отравился ягодами.
                        # Аптечки не оказалось, пришлось возвращаться в город.
                        print_and_save('Перед глазами все плывет. Аптечки в рюкзаке не оказалось, так что '
                                       'нужно возвращаться в город и обратиться к врачу.')
                        print_and_save('Конец!')

                else:  # Если были выбраны орехи или яблоки
                    # Концовка: Студент восстановил силы и успешно завершил поход.
                    print_and_save('Ты восстановил силы и успешно доходишь до водопада!')
                    print_and_save('Конец!')

    elif route_choice == 'к':  # Если выбрана короткая дорога
        print_and_save('Вокруг тебя высокие деревья, пахнет свежестью, слышен легкий шелест листьев.\n'
                       'Ты наслаждаешься своим маленьким путешествием.')
        print_and_save('Казалось, что все проблемы позади, но ты услышал какой-то треск.\n'
                       'Где-то впереди виднеется большая темная фигура... Это медведь!')

        darts = 2 + ('дротик' in belongings['Рюкзак'])  # прибавление дротика из рюкзака к общему кол-ву, если он есть
        print_and_save('Он поднимается на задние лапы и, кажется, готовится к атаке!\n'
                       'Важно угадать, с какой стороны он собирается напасть.\n'
                       'У тебя есть усыпляющее ружье. Нужно действовать быстро!')
        # Битва с медведем
        while darts:  # Пока дротики не закончились
            direction = input_and_save('Куда выстрелить дротиком? (влево/вправо/прямо): ').lower()
            if direction == 'вправо':  # Если направление атаки угадано правильно
                print_and_save('Есть попадание! Медведя начинает покачивать из стороны в сторону, \n'
                               'пока он не сваливается перед тобой. Медведь уснул!')
                # Концовка: Студент побеждает медведя и успешно завершает поход.
                print_and_save('Ты успешно победил медведя и завершил поход!')
                print_and_save('Конец!')
                break

            elif direction in ('влево', 'прямо'):  # Если направление атаки угадано неверно
                print_and_save('Промах! Медведь приближается...')
                darts -= 1  # Вычитаем использованный дротик из общего кол-ва
                if darts == 0:  # Если дротики закончились
                    # Концовка: Медведь не побежден, студенту приходиться сбегать. Поход сорвался
                    print_and_save('У тебя не осталось дротиков! Ты с трудом убегаешь от зверя, но поход окончен.')
                    print_and_save('Конец!')
            else:
                print_and_save('Введи направление верно!')
else:
    # Концовка: Слишком много неприятностей. Студент возвращается домой.
    print_and_save('Слишком много неприятностей на пути! Ты вынужден вернуться назад.')
    print_and_save('Конец!')
```
